## Week - 02: Data Visualization through different means(Histogram, Scatter plots, Heatmaps, etc)


### 1). Using Crop Recommendation Dataset:

```python
features = crop.columns.to_list()
features.remove("label")
print(features)
```

```python
# Visualizing the distribution of each feature:

# Descriptive statistics for numerical columns in Crop Recommendation Dataset:
crop_stats = crop.describe()

# Plotting distributions of numerical features in the Crop Recommendation dataset:
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Feature Distributions in Crop Recommendation Dataset\n", fontsize=14, fontweight='bold')

plt.style.use("ggplot")                   # --> For better visualization

# Plotting the histograms with KDE:
sns.histplot(crop["N"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[0, 0]).set(title="Nitrogen (N)")
sns.histplot(crop["P"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[0, 1]).set(title="Phosphorus (P)")
sns.histplot(crop["K"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[0, 2]).set(title="Potassium (K)")
sns.histplot(crop["temperature"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[1, 0]).set(title="Temperature")
sns.histplot(crop["ph"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[1, 1]).set(title="pH Level")
sns.histplot(crop["rainfall"], bins=20, kde=True, color=plt.cm.viridis(0.5), edgecolor='black', ax=axes[1, 2]).set(title="Rainfall")

# Adjusting layout and display the plot:
plt.tight_layout()
plt.show()
```

```python
# Plotting scatter plot:

# Customizations for the scatter plot:
fig, ax = plt.subplots(3, 3, figsize=(15, 15))

fig.suptitle("Scatter Plots for Data Relationship Visualization\n", fontsize=14, fontweight='bold')

for i, subplot in zip(features, ax.flatten()):
    # Creating the scatter plot with customizations:
    sns.scatterplot(x=i, y='label', data=crop, ax=subplot, color='skyblue', s=100, marker='o', edgecolor='black')
    
    # Customizing the titles and axis labels:
    subplot.set_title(f'Scatter of {i} vs Label', fontsize=12)
    subplot.set_xlabel(i, fontsize=12)
    subplot.set_ylabel('Label', fontsize=12)

    # Customizing grid and background:
    subplot.grid(True, linestyle='--', alpha=0.6)
    subplot.set_facecolor('#f0f0f0')

# Adjusting layout to avoid overlapping:
plt.tight_layout()

plt.show()
```

```python
# Finding the outliers in the dataset by plotting the "boxplot":

# Creating a figure with customized appearance:
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle("Boxplots for Detecting Outliers\n", fontsize=16, fontweight='bold')
print("")

# Customizing each subplot:
for i, subplot in zip(features, ax.flatten()):
    sns.boxplot(x=crop[i], ax=subplot, color='skyblue', width=0.6, fliersize=6, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 6})
    
    # Customizing titles and labels:
    subplot.set_title(f"Boxplot of {i}", fontsize=12)
    subplot.set_xlabel(i, fontsize=12)
    
    # Customizing grid and background:
    subplot.grid(True, linestyle='--', alpha=0.6)
    subplot.set_facecolor('#f0f0f0')

plt.tight_layout()
plt.show()
```

```python
# Correlation of the features:

# Selecting only numerical columns from the dataset:
num_cols = crop.select_dtypes(include=[np.number])          # --> It'll filter out the columns with numerical data types only
num_cols                                                    # --> It'll display the selected numerical columns

# Calculating the correlation matrix:
corr = num_cols.corr()                                      # --> Computes the correlation between numerical features in the dataset
corr                                                        # --> Display the correlation matrix
```

```python
# Visualizing the correlation with Heatmap:

plt.figure(figsize=(10, 6))                                    # --> Set the figure size for better readability

sns.heatmap(corr, annot=True, fmt = '0.2f', cmap='coolwarm')  # --> This'll generate the heatmap with correlation values
plt.title("Feature Correlation Heatmap for Crop Recommendation\n", fontsize=16, fontweight='bold', pad=12)

plt.show()
```
